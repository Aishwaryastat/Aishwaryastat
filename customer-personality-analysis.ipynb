Customer Personality Analysis is a detailed analysis of a company’s ideal customers. It helps a business to better understand its customers and makes it easier for them to modify products according to the specific needs, behaviors and concerns of different types of customers.
Customer personality analysis helps a business to modify its product based on its target customers from different types of customer segments. For example, instead of spending money to market a new product to every customer in the company’s database, a company can analyze which customer segment is most likely to buy the product and then market the product only on that particular segment.
Attributes

People
•	ID: Customer's unique identifier

•	Year_Birth: Customer's birth year

•	Education: Customer's education level

•	Marital_Status: Customer's marital status

•	Income: Customer's yearly household income

•	Kidhome: Number of children in customer's household

•	Teenhome: Number of teenagers in customer's household

•	Dt_Customer: Date of customer's enrollment with the company

•	Recency: Number of days since customer's last purchase

•	Complain: 1 if the customer complained in the last 2 years, 0 otherwise
Products

•	MntWines: Amount spent on wine in last 2 years

•	MntFruits: Amount spent on fruits in last 2 years

•	MntMeatProducts: Amount spent on meat in last 2 years

•	MntFishProducts: Amount spent on fish in last 2 years

•	MntSweetProducts: Amount spent on sweets in last 2 years

•	MntGoldProds: Amount spent on gold in last 2 years
Promotion

•	NumDealsPurchases: Number of purchases made with a discount

•	AcceptedCmp1: 1 if customer accepted the offer in the 1st campaign, 0 
otherwise

•	AcceptedCmp2: 1 if customer accepted the offer in the 2nd campaign, 0 otherwise

•	AcceptedCmp3: 1 if customer accepted the offer in the 3rd campaign, 0 otherwise

•	AcceptedCmp4: 1 if customer accepted the offer in the 4th campaign, 0 otherwise

•	AcceptedCmp5: 1 if customer accepted the offer in the 5th campaign, 0 otherwise

•	Response: 1 if customer accepted the offer in the last campaign, 0 otherwise
Place

•	NumWebPurchases: Number of purchases made through the company’s website

•	NumCatalogPurchases: Number of purchases made using a catalogue

•	NumStorePurchases: Number of purchases made directly in stores

•	NumWebVisitsMonth: Number of visits to company’s website in the last month

# Feature Engineering
 

Create a "Age" feature to indicate the customer's age range.

```python
from datetime import date
todays_date= date.today()
print("Current date:", todays_date)
print("current year:", todays_date.year)
```

```python
df["Age"]=todays_date.year-df["Year_Birth"]
df["Age"]
```

Make a feature that will extract the customer's overall spending on all products.

```python
#Total Spending 
df["Total_spend"]=df["MntFishProducts"]+df["MntFruits"]+df["MntGoldProds"]+df["MntMeatProducts"]+df["MntSweetProducts"]+df["MntWines"]
df["Total_spend"]
```

Replace some of the values in "Education" and "Marital_Status" to produce data that is relevant.

```python
df["Education"]=df["Education"].replace({"Basic":"Undergraduate","2n Cycle":"Undergraduate","Graduation":"Graduation",
                                         "PhD":"Postgraduate","Master":"Postgraduate"})
df["Education"].value_counts()
 
```

```python
df["Living_Status"]=df["Marital_Status"].replace({"Married":"Partner","Together":"Partner","Single":"Alone",
                                                "Divorced":"Alone","Widow":"Alone","Absurd":"Alone","YOLO":"Alone"})
df["Living_Status"].value_counts()
 
```

Count all children in the family, including teenagers and little children. and develop fresh features for the entire children.

```python
df["child"]=df["Kidhome"]+df["Teenhome"]
df["child"]
```

Add a new feature called family_size that includes both parents and children.

```python
df["family_size"]=df["Living_Status"].replace({"Partner":2,"Alone":1})+df["child"]
df["family_size"]
```

```python
# Drop features
df.drop(["ID","Year_Birth","Marital_Status","Dt_Customer","Z_CostContact","Z_Revenue"],axis=1,inplace=True)
```

Find out some outliers.

```python
print(sns.boxplot(df["Income"]))
```

We can observe that there is an anomaly in the income. We earn more than $600,000 annually.

```python
print(sns.boxplot(df["Age"]))
```

As you can see, a customer can have a maximum of 130 years. That is absurd, yet the data is outdated, and my age calculation starts in 2023. 
Therefore, we will discard the data for individuals older than 85.


```python
df=df[df["Age"]<85]
df=df[df["Income"]<600000]
df.describe()
      
```

Heatmap to identify correlation

```python
correlation_maxtrix=df.corr()
plt.figure(figsize=(20,20))
sns.heatmap(correlation_maxtrix,annot=True,cmap="coolwarm",center=0)
```


# Data Preprocessing 
 

In order to achieve better clustering results, we aim to standardise the data during data processing and perform label encoding with object data type features.


```python
a=(df.dtypes=="object")
obj=list(a[a].index)
obj
```

```python
le=LabelEncoder()
for i in obj:
    df[i]=df[[i]].apply(le.fit_transform)
    
df.info()
```

```python
new_data=df.copy()
col=["AcceptedCmp3","AcceptedCmp4","AcceptedCmp5","AcceptedCmp1","AcceptedCmp2","Complain","Response"]
new_data.drop(col,axis=1)

sd=StandardScaler()
sd.fit(new_data)
scaled_data=pd.DataFrame(sd.transform(new_data),columns=new_data.columns)
scaled_data
```

# Dimensionality Reduction 
 

We have many features in the data. The difficulty of the work increases with the number of features offered. Not all features contribute an equal amount of information. Many characteristics are linked and hence redundant. 
Therefore, we attempt to use Principle Component Analysis (PCA) to decrease the feature. 

```python
pca=PCA()
pca_value=pca.fit_transform(scaled_data)
pca_value
```

```python
pca.explained_variance_
```

```python
var=pca.explained_variance_ratio_
var
```

```python
var_per=np.cumsum(np.round(var,decimals=4)*100)
var_per
```

```python
pca.components_
```

```python
# Variance Percentage 
plt.plot(var_per,color="red")
```

```python
pca_value[0:10]
```

In order to generate clusters, we have chosen six key components and created a final dataframe.

```python
final_df=pd.DataFrame(pca_value[:,0:6],columns=["pca1","pca2","pca3","pca4","pca5","pca6"])
final_df 
```

```python
sns.pairplot(final_df)
```


# Clustering 
 

## Hierarchical Clustering

```python
plt.figure(figsize=(18,8))
dendrogram= sch.dendrogram(sch.linkage(final_df,method="average"))
```

From above dendrogram, we suggest 4 cluster to create.

```python
hc= AgglomerativeClustering(n_clusters=4,affinity="euclidean",linkage="average")
```

```python
predict_cluster_hc = hc.fit_predict(final_df)
predict_cluster_hc
```

```python
cluster_hc=pd.DataFrame(predict_cluster_hc,columns=["Cluster_hc"])
cluster_data_hc=pd.concat([final_df,cluster_hc],axis=1)
cluster_data_hc
```

# KMEAN


```python
# To determine the value of K, we use Yellowbrick Library. 
print('Elbow Method to determine the number of clusters to be formed:')
Elbow_M = KElbowVisualizer(KMeans(), k=10)
Elbow_M.fit(final_df)
Elbow_M.show()
```

Above Elbow visualizer suggest to create 5 clusters. 

```python
cluster=KMeans(n_clusters=5).fit(final_df)
cluster.labels_
```

```python
cluster_kmean=pd.DataFrame(cluster.labels_,columns=["Cluster_kmean"])
cluster_data_kmean=pd.concat([final_df,cluster_kmean],axis=1)
cluster_data_kmean
```
